```python
import face_recognition
import os
import matplotlib.pyplot as plt
import pickle
import cv2
```

```python
def read_model(filename, path):
    with open(os.path.join(path, filename), 'rb') as in_name:
        model = pickle.load(in_name)
        
        return model
```

```python
model = read_model('deuxiemeModel.pkl','C:/Users/Ibrahim/Desktop/projetVA/models')
```

```python
test = 'C:/Users/Ibrahim/Desktop/projetVA/testImages/mj.jpg'
test_image = face_recognition.load_image_file(test)
plt.imshow(test_image)

#Find all the faces in the test image using the default HOG-based model
face_locations = face_recognition.face_locations(test_image)
no = len(face_locations)
print("Number of faces detected: ", no)
```

```python

for k in range(no):
    print(k)
    print(face_locations[k])
    test_image_enc = face_recognition.face_encodings(test_image)[k]
    name= model.predict([test_image_enc])
    prediction = model.predict_proba([test_image_enc])
    print(max(prediction[0]))
    #print(prediction)
    print(*name)
```

```python
cascPath = 'C:/Users/Ibrahim/Desktop/projetVA/data/visage.xml'
faceCascade = cv2.CascadeClassifier(cascPath)
faces = faceCascade.detectMultiScale(
    test_image,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30)
)

print(len(faces))
k=0

for (x, y, w, h) in faces:
    if(k<no):
        test_image_enc = face_recognition.face_encodings(test_image)[k]
        name= model.predict([test_image_enc])
        probability = model.predict_proba([test_image_enc])
        m=max(probability[0])
        k+=1
        if(m < 0.4):
            name=['inconnu']
        cv2.rectangle(test_image, (x, y), (x+w, y+h), (0, 255, 0), 2)
        font = cv2.FONT_ITALIC

        org = (50, 50) 

        fontScale = 1

        color = (255, 0, 0) 
        thickness = 2
        if(k%2==0):
            cv2.putText(test_image, *name, (x,y), font,  fontScale, color, thickness, cv2.LINE_AA)
        else :
            cv2.putText(test_image, *name, (x,y+90), font,  fontScale, color, thickness, cv2.LINE_AA)
    
```
